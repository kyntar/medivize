```python
# @title <h1>Notebook A-Z: Persiapan Data & Pelatihan YOLO di Google Drive</h1>
# @markdown ### Dibuat oleh Gemini (Diperbarui & Diperbaiki)
# @markdown ---
# @markdown **Alur Kerja Notebook:**
# @markdown 1.  **Hubungkan & Siapkan Google Drive:** Membuat folder proyek utama.
# @markdown 2.  **Unduh, Proses & Validasi Dataset:**
# @markdown     - Mengekstrak, membersihkan nama label, dan membuat Bounding Box dengan **logika penggabungan kontur untuk menangani detail internal objek**.
# @markdown     - **Validasi:** Menampilkan 5 sampel acak **dengan bounding box** dari setiap set.
# @markdown 3.  **Latih Model YOLO dengan Augmentasi:**
# @markdown     - Memulai pelatihan untuk menghasilkan model `best.pt`.
# @markdown 4.  **Uji Coba Inferensi (Opsional):**
# @markdown     - Menjalankan prediksi pada gambar acak dari set tes.
# @markdown 5.  **Ekspor & Unduh Artefak Pelatihan:**
# @markdown     - Mengemas model **`best.pt` (PyTorch)**, metadata, dan `labels.json` ke dalam satu file ZIP untuk diunduh.
# @markdown ---
# @markdown **Petunjuk:** Jalankan sel ini dari atas ke bawah.

# ==============================================================================
# LANGKAH 1: PENGATURAN AWAL & HUBUNGKAN KE GOOGLE DRIVE
# ==============================================================================
print("="*80)
print("LANGKAH 1: PENGATURAN AWAL & HUBUNGKAN KE GOOGLE DRIVE")
print("="*80)

import os
import shutil
import glob
import random
import yaml
import cv2
import numpy as np
import re
import zipfile
import json
from tqdm.notebook import tqdm
from google.colab import drive, files
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# @markdown ### üìÇ Tentukan Nama Proyek Anda
PROJECT_NAME = "Medivize_v34"  #@param {type:"string"}

print("Menghubungkan ke Google Drive...")
drive.mount('/content/drive')

DRIVE_PROJECT_PATH = f"/content/drive/MyDrive/{PROJECT_NAME}"
os.makedirs(DRIVE_PROJECT_PATH, exist_ok=True)
print(f"‚úÖ Google Drive terhubung. Folder proyek diatur ke: {DRIVE_PROJECT_PATH}")

# Pindah ke direktori kerja utama di Google Drive
%cd {DRIVE_PROJECT_PATH}
print(f"Direktori kerja saat ini: {os.getcwd()}")


# @markdown ### ‚öôÔ∏è Instal Pustaka yang Diperlukan
print("\n‚öôÔ∏è Menginstal pustaka yang diperlukan (ultralytics, opencv, dll)...")
!pip install -q ultralytics opencv-python Pillow numpy tqdm pyyaml matplotlib
print("‚úÖ Pustaka berhasil diinstal.")

# ==============================================================================
# LANGKAH 2: PERSIAPAN DATASET & VALIDASI HASIL
# ==============================================================================
print("\n" + "="*80)
print("LANGKAH 2: PERSIAPAN DATASET (DENGAN BOUNDING BOX OTOMATIS YANG DISEMPURNAKAN)")
print("="*80)

# --- Definisi Nama dan Path ---
DATASET_DIR = f"{PROJECT_NAME}_YOLO"
ZIP_FILE = "dataset.zip"
DATASET_URL = "https://prod-dcd-datasets-cache-zipfiles.s3.eu-west-1.amazonaws.com/bjy2svvmn8-1.zip"
YAML_PATH = os.path.join(DATASET_DIR, "data.yaml")
SOURCE_DIR_PATH = None

# --- [PERBAIKAN KRUSIAL] Fungsi Bounding Box yang menggabungkan semua bagian objek ---
def generate_automatic_bounding_box(image_path):
    """
    Menghasilkan bounding box yang ketat dengan menggabungkan semua kontur relevan.
    Metode ini tangguh terhadap objek dengan teks atau detail internal yang kuat,
    serta objek yang miring.
    """
    try:
        image = cv2.imread(image_path)
        if image is None: return None
        orig_h, orig_w = image.shape[:2]

        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        blurred = cv2.GaussianBlur(gray, (5, 5), 0)
        edged = cv2.Canny(blurred, 50, 150)
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5)) # Kernel sedikit lebih besar
        dilated = cv2.dilate(edged, kernel, iterations=2)
        contours, _ = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        if contours:
            # --- PERBAIKAN UTAMA: Gabungkan semua kontur yang signifikan ---
            all_contours = np.vstack([contour for contour in contours if cv2.contourArea(contour) > 100])
            if all_contours.size == 0: return None

            # Dapatkan bounding box dari gabungan semua kontur
            x, y, w, h = cv2.boundingRect(all_contours)

            # Konversi ke format YOLO
            x_center = (x + w / 2) / orig_w
            y_center = (y + h / 2) / orig_h
            width = w / orig_w
            height = h / orig_h

            if width <= 0.01 or height <= 0.01: return None
            return (x_center, y_center, width, height)

    except Exception as e:
        if "ValueError: need at least one array to stack" in str(e):
             return None # Tidak ada kontur signifikan yang ditemukan
        print(f"Error saat memproses {os.path.basename(image_path)} untuk bbox: {e}")
    return None


# --- Fungsi untuk menampilkan sampel dengan Bounding Box ---
def display_samples_with_boxes(image_folder, label_folder, class_names, title, num_samples=5):
    """Fungsi untuk menampilkan sampel gambar secara acak dari sebuah folder beserta bounding box-nya."""
    print(f"\n--- {title} ---")
    if not os.path.exists(image_folder):
        print(f"‚ùå Peringatan: Direktori gambar tidak ditemukan di '{image_folder}'.")
        return

    image_files = glob.glob(os.path.join(image_folder, '*.jpg')) + glob.glob(os.path.join(image_folder, '*.png'))
    if not image_files:
        print(f"Tidak ada gambar yang ditemukan di '{image_folder}'.")
        return

    num_to_sample = min(num_samples, len(image_files))
    if num_to_sample == 0: return

    random_samples = random.sample(image_files, num_to_sample)
    plt.figure(figsize=(25, 6))
    plt.suptitle(title, fontsize=16)

    for i, img_path in enumerate(random_samples):
        plt.subplot(1, num_to_sample, i + 1)
        img = cv2.imread(img_path)
        label_filename = os.path.splitext(os.path.basename(img_path))[0] + ".txt"
        label_path = os.path.join(label_folder, label_filename)

        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                line = f.readline().strip().split()
                class_id, x_c, y_c, w, h = map(float, line)
            label_text = class_names[int(class_id)]
            img_h, img_w = img.shape[:2]
            x1, y1 = int((x_c - w/2) * img_w), int((y_c - h/2) * img_h)
            x2, y2 = int((x_c + w/2) * img_w), int((y_c + h/2) * img_h)
            cv2.rectangle(img, (x1, y1), (x2, y2), (36, 255, 12), 2)
            cv2.putText(img, label_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (36, 255, 12), 2)

        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img_rgb)
        plt.title(os.path.basename(img_path), fontsize=8)
        plt.axis('off')
    plt.show()

# --- LOGIKA UTAMA PERSIAPAN DATA ---
if os.path.exists(YAML_PATH):
    print(f"‚úÖ Folder '{DATASET_DIR}' dengan 'data.yaml' sudah ada.")
    with open(YAML_PATH, 'r') as f:
        data_yaml = yaml.safe_load(f)
        final_class_names = data_yaml['names']
    print("üëç Melewati seluruh proses persiapan data.")

    paths = { "train_images": os.path.join(DATASET_DIR, "train", "images"), "train_labels": os.path.join(DATASET_DIR, "train", "labels"), "valid_images": os.path.join(DATASET_DIR, "valid", "images"), "valid_labels": os.path.join(DATASET_DIR, "valid", "labels"), "test_images": os.path.join(DATASET_DIR, "test", "images"), "test_labels": os.path.join(DATASET_DIR, "test", "labels")}
    display_samples_with_boxes(paths['train_images'], paths['train_labels'], final_class_names, "Contoh dari Set 'Train'")
    display_samples_with_boxes(paths['valid_images'], paths['valid_labels'], final_class_names, "Contoh dari Set 'Valid' (Validasi)")
    display_samples_with_boxes(paths['test_images'], paths['test_labels'], final_class_names, "Contoh dari Set 'Test'")

else:
    print(f"üìÇ Hasil akhir '{YAML_PATH}' belum ditemukan. Memulai proses persiapan data...")
    if os.path.exists(DATASET_DIR):
        shutil.rmtree(DATASET_DIR)

    while True:
        if not os.path.exists(ZIP_FILE):
            print("   -> Mengunduh dataset...")
            !wget -q --show-progress -O {ZIP_FILE} {DATASET_URL}
        print("\n   -> Memeriksa integritas file ZIP...")
        try:
            with zipfile.ZipFile(ZIP_FILE, 'r') as zf:
                if zf.testzip() is not None: raise zipfile.BadZipFile("File korup.")
            print("   -> Integritas ZIP bagus. Melanjutkan ke ekstraksi...")
            break
        except zipfile.BadZipFile:
            print(f"   -> ‚ùå Peringatan: File '{ZIP_FILE}' rusak. Menghapus dan mengunduh ulang...")
            os.remove(ZIP_FILE)

    print("\n   -> Mengekstrak file ZIP...")
    with zipfile.ZipFile(ZIP_FILE, 'r') as zf:
        all_paths = zf.namelist()
        if all_paths: SOURCE_DIR_PATH = all_paths[0].split('/')[0]
        for member in tqdm(zf.infolist(), desc='   Mengekstrak '):
            zf.extract(member, '.')
    if not SOURCE_DIR_PATH or not os.path.exists(SOURCE_DIR_PATH):
        raise FileNotFoundError("Gagal menemukan direktori hasil ekstraksi.")

    print("\n‚öôÔ∏è Memulai proses persiapan data...")
    paths = { "train_images": os.path.join(DATASET_DIR, "train", "images"), "train_labels": os.path.join(DATASET_DIR, "train", "labels"), "valid_images": os.path.join(DATASET_DIR, "valid", "images"), "valid_labels": os.path.join(DATASET_DIR, "valid", "labels"), "test_images": os.path.join(DATASET_DIR, "test", "images"), "test_labels": os.path.join(DATASET_DIR, "test", "labels")}
    for p in paths.values(): os.makedirs(p, exist_ok=True)

    def clean_label_name(name):
        match = re.match(r'(.+?)(?=\s+\d+\s|\s+nan\s|$)', name)
        return match.group(1).strip() if match else name.strip()

    raw_class_names = sorted([d for d in os.listdir(SOURCE_DIR_PATH) if os.path.isdir(os.path.join(SOURCE_DIR_PATH, d))])
    raw_to_cleaned_map = {raw: clean_label_name(raw) for raw in raw_class_names}
    final_class_names = sorted(list(set(raw_to_cleaned_map.values())))
    class_map = {name: i for i, name in enumerate(final_class_names)}

    print("\n" + "="*50)
    print("üè∑Ô∏è   LABEL AKHIR (SETELAH DIBERSIHKAN) UNTUK PELATIHAN")
    print(f"Total kelas unik: {len(final_class_names)}")
    for i, name in enumerate(final_class_names): print(f"   {i}: {name}")
    print("="*50 + "\n")

    all_files = []
    for raw_name in raw_class_names:
        class_id = class_map[clean_label_name(raw_name)]
        class_dir = os.path.join(SOURCE_DIR_PATH, raw_name)
        for img_name in os.listdir(class_dir):
            if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                all_files.append({"source_path": os.path.join(class_dir, img_name), "class_id": class_id, "original_name": img_name})

    random.shuffle(all_files)
    train_idx, valid_idx = int(len(all_files)*0.7), int(len(all_files)*0.9)
    splits = {"train": all_files[:train_idx], "valid": all_files[train_idx:valid_idx], "test": all_files[valid_idx:]}

    for split_name, file_list in splits.items():
        print(f"\n   -> Memproses set '{split_name}'...")
        for file_info in tqdm(file_list, desc=f"   Memproses {split_name}"):
            try:
                dest_img_path = os.path.join(paths[f"{split_name}_images"], file_info["original_name"])
                shutil.copy(file_info["source_path"], dest_img_path)
                bbox = generate_automatic_bounding_box(dest_img_path) or (0.5, 0.5, 1.0, 1.0)
                label_name = os.path.splitext(file_info["original_name"])[0] + ".txt"
                with open(os.path.join(paths[f"{split_name}_labels"], label_name), 'w') as f:
                    f.write(f"{file_info['class_id']} {bbox[0]:.6f} {bbox[1]:.6f} {bbox[2]:.6f} {bbox[3]:.6f}")
            except Exception as e:
                print(f"Gagal memproses file {file_info['original_name']}: {e}")

    IMG_SIZE = 640
    print("\n   -> Mengubah ukuran gambar ke 640x640...")
    for split_name in splits.keys():
        for img_name in tqdm(os.listdir(paths[f"{split_name}_images"]), desc=f"   Mengubah ukuran {split_name}"):
            try:
                img_path = os.path.join(paths[f"{split_name}_images"], img_name)
                img = cv2.imread(img_path)
                if img is not None:
                    cv2.imwrite(img_path, cv2.resize(img, (IMG_SIZE, IMG_SIZE)))
            except Exception as e:
                print(f"Gagal mengubah ukuran {img_name}: {e}")

    yaml_content = {'path': os.path.abspath(DATASET_DIR), 'train': 'train/images', 'val': 'valid/images', 'test': 'test/images', 'nc': len(final_class_names), 'names': final_class_names}
    with open(YAML_PATH, 'w') as f: yaml.dump(yaml_content, f, sort_keys=False)
    shutil.rmtree(SOURCE_DIR_PATH)
    print("\n‚úÖ Semua proses persiapan data telah selesai.")

    # --- VALIDASI HASIL PELABELAN ---
    print("\n" + "="*80)
    print("VALIDASI HASIL PELABELAN: MENAMPILKAN 5 CONTOH DARI SETIAP SET")
    print("="*80)
    display_samples_with_boxes(paths['train_images'], paths['train_labels'], final_class_names, "Contoh dari Set 'Train'")
    display_samples_with_boxes(paths['valid_images'], paths['valid_labels'], final_class_names, "Contoh dari Set 'Valid' (Validasi)")
    display_samples_with_boxes(paths['test_images'], paths['test_labels'], final_class_names, "Contoh dari Set 'Test'")

# ==============================================================================
# LANGKAH 3: PELATIHAN MODEL YOLO
# ==============================================================================
print("\n" + "="*80)
print("LANGKAH 3: PELATIHAN MODEL YOLO DENGAN AUGMENTASI DATA")
print("="*80)

from ultralytics import YOLO

# @markdown ### üîÑ Lanjutkan Pelatihan (Opsional)
RESUME_TRAINING = False  #@param {type:"boolean"}

model = YOLO("yolov8n.pt")
dataset_yaml_path = os.path.join(DRIVE_PROJECT_PATH, YAML_PATH)
print(f"\nFile konfigurasi dataset yang akan digunakan: {dataset_yaml_path}\n")

if RESUME_TRAINING:
    print("üîÑ Mencoba melanjutkan pelatihan dari checkpoint terakhir...")
    list_of_runs = glob.glob(f'{DRIVE_PROJECT_PATH}/{PROJECT_NAME}_YOLOv8n_augmented_results*/weights/last.pt')
    if list_of_runs:
        latest_run = max(list_of_runs, key=os.path.getctime)
        print(f"Melanjutkan dari: {latest_run}")
        model = YOLO(latest_run)
    else:
        print("‚ùå Tidak ada checkpoint terakhir yang ditemukan. Memulai pelatihan baru.")
        RESUME_TRAINING = False
else:
    print("üöÄ Memulai sesi pelatihan baru...")

results = model.train(
    data=dataset_yaml_path, epochs=100, imgsz=640, project=DRIVE_PROJECT_PATH,
    name=f'{PROJECT_NAME}_YOLOv8n_augmented_results', exist_ok=True,
    fliplr=0.5, mosaic=1.0, mixup=0.1, scale=0.5, translate=0.1, patience=20,
    resume=RESUME_TRAINING
)
print(f"\nüéâ Pelatihan selesai. Hasil disimpan di: {results.save_dir}")

# ==============================================================================
# LANGKAH 4: UJI COBA INFERENSI
# ==============================================================================
print("\n" + "="*80)
print("LANGKAH 4: UJI COBA INFERENSI PADA GAMBAR TES")
print("="*80)

try:
    if 'results' in locals() and hasattr(results, 'save_dir'):
        model_best_path = os.path.join(results.save_dir, 'weights/best.pt')
    else:
        run_folder_pattern = f'{PROJECT_NAME}_YOLOv8n_augmented_results*'
        list_of_runs = glob.glob(os.path.join(DRIVE_PROJECT_PATH, run_folder_pattern))
        latest_run_dir = max(list_of_runs, key=os.path.getctime)
        model_best_path = os.path.join(latest_run_dir, 'weights/best.pt')

    if not os.path.exists(model_best_path):
        raise FileNotFoundError(f"File model 'best.pt' tidak ditemukan.")

    model_inference = YOLO(model_best_path)
    print(f"‚úÖ Model terbaik dimuat dari: {model_best_path}")
    test_images_path = os.path.join(DRIVE_PROJECT_PATH, DATASET_DIR, 'test/images')
    all_test_images = glob.glob(os.path.join(test_images_path, '*.*'))

    if all_test_images:
        random_image_path = random.choice(all_test_images)
        print(f"\nüîé Melakukan inferensi pada gambar acak: {os.path.basename(random_image_path)}")
        inference_results = model_inference.predict(source=random_image_path)
        result_image_with_boxes = inference_results[0].plot()
        print("\nüñºÔ∏è Hasil Deteksi:")
        cv2_imshow(result_image_with_boxes)
    else:
        print(f"‚ùå Tidak ada gambar yang ditemukan di folder tes: {test_images_path}")
except Exception as e:
    print(f"‚ùå Error saat inferensi: {e}. Pastikan Langkah 3 telah dijalankan.")

# ==============================================================================
# LANGKAH 5: EKSPOR & UNDUH ARTEFAK PELATIHAN (.PT)
# ==============================================================================
print("\n" + "="*80)
print("LANGKAH 5: EKSPOR & UNDUH ARTEFAK PELATIHAN (.PT)")
print("="*80)

try:
    export_base_path = os.path.join(DRIVE_PROJECT_PATH, f"{PROJECT_NAME}_Training_Artifacts")
    os.makedirs(export_base_path, exist_ok=True)

    # --- [PERUBAHAN] Cari dan salin model best.pt ---
    # Gunakan path yang sudah ditemukan di Langkah 4 jika ada
    if 'model_best_path' in locals() and os.path.exists(model_best_path):
         best_pt_source_path = model_best_path
    else:
        # Fallback jika notebook dijalankan ulang setelah training
        run_folder_pattern = f'{PROJECT_NAME}_YOLOv8n_augmented_results*'
        list_of_runs = glob.glob(os.path.join(DRIVE_PROJECT_PATH, run_folder_pattern))
        latest_run_dir = max(list_of_runs, key=os.path.getctime)
        best_pt_source_path = os.path.join(latest_run_dir, 'weights/best.pt')

    if os.path.exists(best_pt_source_path):
        print(f"   -> Menyalin model terbaik: {os.path.basename(best_pt_source_path)}")
        shutil.copy(best_pt_source_path, os.path.join(export_base_path, "best.pt"))
    else:
        print(f"   -> ‚ùå Peringatan: Model 'best.pt' tidak ditemukan.")

    # Salin file data.yaml
    dataset_yaml_path = os.path.join(DRIVE_PROJECT_PATH, YAML_PATH)
    if os.path.exists(dataset_yaml_path):
        print(f"   -> Menyalin file konfigurasi: {os.path.basename(dataset_yaml_path)}")
        shutil.copy(dataset_yaml_path, os.path.join(export_base_path, "data.yaml"))

    # Buat labels.json
    with open(dataset_yaml_path, 'r') as f:
        class_names_from_yaml = yaml.safe_load(f)['names']
    labels_dict = {"labels": class_names_from_yaml}
    with open(os.path.join(export_base_path, "labels.json"), 'w') as f_json:
        json.dump(labels_dict, f_json, indent=4)
    print("   -> Berhasil membuat labels.json.")

    # Buat file ZIP
    archive_name = f"{PROJECT_NAME}_Training_Artifacts"
    shutil.make_archive(os.path.join(DRIVE_PROJECT_PATH, archive_name), 'zip', export_base_path)
    print(f"‚úÖ File '{archive_name}.zip' berhasil dibuat di Google Drive Anda.")

    # Unduh file ZIP ke komputer lokal
    print("\n--- Memulai Proses Pengunduhan ---")
    files.download(os.path.join(DRIVE_PROJECT_PATH, f"{archive_name}.zip"))

except Exception as e:
    print(f"‚ùå Terjadi kesalahan saat ekspor: {e}")
```
